### Инструкции для функций:

#### `__init__`

В `__init__` приходит кучка параметров, распишем что есть что:

 - `base_model_class` - класс базовой модели нашего бустинга
 - `base_model_params` - словарь с гиперпараметрами для базовой модели
 - `n_estimators` - какое количество базовых моделей нужно обучить
 - `learning_rate` - темп обучения, должен быть из полуинтервала $(0, 1]$
 - `subsample` - доля объектов, на которой будет обучаться базовая модель (какую часть составляет бутстрапная выборка от исходной обучающей)
 - `early_stopping_rounds` - число итераций, после которых при отсутствии улучшения качества на валидационной выборке обучение останавливается
 - `plot` - строить ли после обучения всех базовых моделей график с качеством

#### `fit`

В `fit` приходит две выборки, обучающая и валидационная. На обучающей мы обучаем новые базовые модели, на валидационной считаем качество для ранней остановки (если это предусматривают параметры).

Сначала нам нужно сделать какую-то нулевую модель, сделать предсказания для обучающей и валидационной выборок (в шаблоне это нулевая модель, соответственно предсказания это просто `np.zeros`). После этого нужно обучить `n_estimators` базовых моделей (как и на что обучаются базовые модели смотрите в лекциях и семинарах). После каждой обученной базовой модели мы должны обновить текущие предсказания, посчитать ошибку на обучающей и валидационной выборках (используем `loss_fn` для этого), проверить на раннюю остановку.

После всего цикла обучения надо нарисовать график (если `plot`).


#### `fit_new_base_model`

В `fit_new_base_model` приходит обучающая выборка (целиком) и текущие предсказания для неё. Мы должны сгенерировать бутстрап выборку для обучения базовой модели и обучить базовую модель. После обучения модели запускаем поиск оптимальной гаммы, добавляем новую модель и гамму (не забываем про темп обучения) в соответствующие списки.

#### `predict_proba`

В `predict_proba` приходит выборка, нужно предсказать вероятности для неё. Суммируем предсказания базовых моделей на этой выборке (не забываем про гаммы) и накидываем сигмоиду.

## Задание 2. Обучение градиентного бустинга (1 балл)

Оцените качество на тестовой выборке вашей имплементации бустинга для различной максимальной глубины решающего дерева в качестве базовой модели. Здесь и далее мы будем использовать метрику ROC-AUC.

Перебирайте максимальную глубину от 1 до 30 с шагом 2 (остальные параметры бустинга стоит оставить равными по умолчанию). Постройте график зависимости качества на обучающей и тестовой выборке в зависимости от глубины.

````python
results = {}

depths = range(1, 30, 2)

train_scores = []
valid_scores = []
test_scores = []

np.random.seed(111)
for depth in depths:
    boosting = Boosting(base_model_params = {'max_depth' : depth})
    boosting.fit(x_train, y_train, x_valid, y_valid)
    
    train_scores.append(boosting.score(x_train, y_train))
    test_scores.append(boosting.score(x_test, y_test))
````

````python
plt.figure(figsize=(10,8))
sns.lineplot(x=depths, y=train_scores, label='train')
sns.lineplot(x=depths, y=test_scores, label='test')

plt.title('Dependence of the score on depth')
plt.xlabel('Depth')
plt.ylabel('Score')
plt.legend()
plt.show()
````

Найду при каком ```max_depth``` получен лучший скор на тестовой выборке

````python
depths[np.argmax(test_scores)]
````

````python
x.shape
````

**Какая из моделей имеет лучшее качество? Как вы можете это объяснить?**

Согласно графикам выше, лучший результат на тестовой выборке моя имплементация бустинга показывает при ```max_depth = 7```, то есть лучший результат достигается, если использовать в композиции не глубокие деревья (так как признаков 169, то деревья глубины 7 можно считать неглубокими).

Этот результат соответствует нашим теоретическим знаниям. Так как в бустинге каждая следующуя модель в композиции исправляет ошибки предыдущей, то при использовании очень глубоких базовых моделей итоговая композиция может сильно переобучиться. Избежать этого помогает как раз использование простых базовых моделей (деревьев небольшой глубины), имеющих высокое смещение. При достаточном числе базовых моделей композиция всё равно получается достаточно сложной и может приблизать сложные зависимости в данных, то есть смещение снижается, а разброс либо остаётся таким же лио увеличивается. Поэтому неглубоки деревья глуюины 7 и показали лучший результат.

## Задание 3. Подбираем гиперпараметры и ищем лучшую модель (2 балла)

Подберите по валидационной выборке основные гиперпараметры для вашей модели бустинга. Следует подобрать все основные параметры для самого градиентного бустинга и для самих базовых моделей.

Основными парамтрами для решающего дерева являются параметры `max_depth`, `min_samples_leaf` и `min_samples_split`. Их и буду подбирать для решеющего дерева. Для градиентного бустинга подберу параметры `n_estimators`, `learning_rate` и `subsample`.

Для начала сделаю перебор по очень крупной сетке, чтобы уточнить, в кокой окресности искать каждый параметр.

````python
from itertools import product
````

````python
np.random.seed(111)

best_result_1 = 0
best_params_1 = {}
# для max_depth рассмотрю параметры, близкие к хорошей глубине, полученной в предыдущем номере
for max_depth, min_samples_leaf, min_samples_split in product([5, 7, 9], [1, 7, 15], [2, 20, 50]):
    params = {'max_depth' : max_depth, 'min_samples_leaf' : min_samples_leaf, 'min_samples_split' : min_samples_split}
    for n_estimators, learning_rate, subsample in product([10, 20, 40], [0.1, 0.3, 0.6], [0.3, 0.5, 0.7]):
        boosting = Boosting(base_model_params=params,
                            n_estimators=n_estimators,
                            learning_rate=learning_rate,
                            subsample=subsample)
        boosting.fit(x_train, y_train, x_valid, y_valid)
        
        train_score = boosting.score(x_train, y_train)
        valid_score = boosting.score(x_valid, y_valid)
        
        if valid_score > best_result_1:
            best_result_1 = valid_score
            best_params_1 = {'max_depth' : max_depth,
                             'min_samples_leaf' : min_samples_leaf,
                             'min_samples_split' : min_samples_split,
                             'n_estimators' : n_estimators,
                             'learning_rate' : learning_rate,
                             'subsample' : subsample}
        print(f'{max_depth}, {min_samples_leaf}, {min_samples_split}, {n_estimators}, {learning_rate}, {subsample}: train=%.4f' % train_score, f', val=%.4f' % valid_score)
````

````python
print(best_result_1)
print(best_params_1)
````

Теперь возьму лучшие параметры и сделаю ещё один перебор в окрестности этих параметров:

````python
np.random.seed(111)

best_result_2 = 0
best_params_2 = {}
for max_depth, min_samples_leaf, min_samples_split in product([6, 7, 8], [1, 3, 5], [15, 20, 30]):
    params = {'max_depth' : max_depth, 'min_samples_leaf' : min_samples_leaf, 'min_samples_split' : min_samples_split}
    for n_estimators, learning_rate, subsample in product([30, 40, 55], [0.5, 0.6, 0.7], [0.4, 0.5, 0.6]):
        boosting = Boosting(base_model_params=params,
                            n_estimators=n_estimators,
                            learning_rate=learning_rate,
                            subsample=subsample)
        boosting.fit(x_train, y_train, x_valid, y_valid)
        
        train_score = boosting.score(x_train, y_train)
        valid_score = boosting.score(x_valid, y_valid)
        
        if valid_score > best_result_2:
            best_result_2 = valid_score
            best_params_2 = {'max_depth' : max_depth,
                             'min_samples_leaf' : min_samples_leaf,
                             'min_samples_split' : min_samples_split,
                             'n_estimators' : n_estimators,
                             'learning_rate' : learning_rate,
                             'subsample' : subsample}
        print(f'{max_depth}, {min_samples_leaf}, {min_samples_split}, {n_estimators}, {learning_rate}, {subsample}: train=%.4f' % train_score, f', val=%.4f' % valid_score)
````

````python
print(best_result_2)
print(best_params_2)
````

Теперь с такими параметрами для деревьев подберу ещё точнее параметры `n_estimators` и `learning_rate`.

````python
np.random.seed(111)

best_result_3 = 0
best_params_3 = {}
params = {'max_depth' : 6, 'min_samples_leaf' : 5, 'min_samples_split' : 15}
for n_estimators, learning_rate in product([50, 55, 60, 65, 70], [0.65, 0.675, 0.7, 0.725, 0.75]):
    boosting = Boosting(base_model_params=params,
                        n_estimators=n_estimators,
                        learning_rate=learning_rate,
                        subsample=0.6)
    boosting.fit(x_train, y_train, x_valid, y_valid)

    train_score = boosting.score(x_train, y_train)
    valid_score = boosting.score(x_valid, y_valid)

    if valid_score > best_result_3:
        best_result_3 = valid_score
        best_params_3 = {'n_estimators' : n_estimators,
                         'learning_rate' : learning_rate}
    print(f'{n_estimators}, {learning_rate}: train=%.4f' % train_score, f', val=%.4f' % valid_score)
````

````python
print(best_result_3)
print(best_params_3)
````

Обучу итоговую модель с лучшими параметрами (которые получились после второго запуска по сетке):

````python
np.random.seed(111)

params = {'max_depth' : 6, 'min_samples_leaf' : 5, 'min_samples_split' : 15}
boosting = Boosting(base_model_params=params,
                    n_estimators=55,
                    learning_rate=0.7,
                    subsample=0.6)
boosting.fit(x_train, y_train, x_valid, y_valid)

train_score = boosting.score(x_train, y_train)
valid_score = boosting.score(x_valid, y_valid)
test_score = boosting.score(x_test, y_test)
    
print(f'train=%.4f' % train_score, f', val=%.4f' % valid_score, f', test=%.4f' % test_score)
````

Таким образом, удалось поулчить результат 0.9401 на тесте.

Постройте калибровочную кривую для вашей лучшей модели бустинга (из тех, что используют вашу реализацию).
Насколько хорошо бустинг оценивает вероятности? Постройте также калибровочную кривую для логистической регресии, сравните их между собой. Проанализируйте полученные результаты.

Воспользуюсь кодом с семинара №6 для построения калибровочных кривых.

````python
from sklearn.calibration import calibration_curve
````

````python
boosting_pred = boosting.predict_proba(x_test)[:, 1]
xgb_pred = xgb.predict_proba(x_test)[:, 1]
````

Построю две калибровочных кривых. Одна - лучшая модель, полученная с помощью моей имплементации бустинга, вторая - построенная с помощью библиотеки xgboost.

````python
plt.figure(figsize=(7, 7))

boosting_true_prob, boosting_pred_prob = calibration_curve(y_test, boosting_pred, n_bins=15)
xgb_true_prob, xgb_pred_prob = calibration_curve(y_test, xgb_pred, n_bins=15)

plt.plot(boosting_pred_prob, boosting_true_prob, label='Boosting', color='blue')
plt.plot(xgb_pred_prob, xgb_true_prob, label='XGBoost', color='orange')
plt.plot([0, 1], [0, 1], label='Perfect', linestyle='--', color='green')

plt.xlabel('Mean predicted probability')
plt.ylabel('Fraction of positives')
plt.title('Calibration curves')
plt.legend()
plt.show()
````

В принципе, оба алгоритма показали неплохой результат, можно сказать, что бустинг хорошо оценил вероятности, так как калибровочная кривая достаточно близка к идеальной.

Бустинг хорошо оценивает вероятности, потому что деревья в бустинге обучаются на MSE, а MSE удовлетворяет свойству

$$
argmin_{b \in R} E \left[ L(y, b) | x \right] = p(y = +1 | x)
$$

при $n$ (число объектов) стремится к бесконечности, то есть хорошо оценивает вероятности.

````python
from sklearn.linear_model import LogisticRegression
````

````python
lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict_proba(x_test)[:, 1]
````

````python
plt.figure(figsize=(7, 7))

lr_true_prob, lr_pred_prob = calibration_curve(y_test, lr_pred, n_bins=15)

plt.plot(lr_pred_prob, lr_true_prob, label='Boosting', color='blue')
plt.plot([0, 1], [0, 1], label='Perfect', linestyle='--', color='green')

plt.xlabel('Mean predicted probability')
plt.ylabel('Fraction of positives')
plt.title('Calibration curves')
plt.legend()
plt.show()
````

Логистическая регрессия хорошо оценивает вероятности, как жто и должно быть.

Построю гистограмму важностей признаков для бустинга и для логистической регрессии. Для бустинга реализую метод `feature_importances_`, согласно заданию, а для логистической регрессии, возьму модули весов, отнормирую в пространстве $l_1$ и буду считать это важностью признаков.

````python
fig, axes = plt.subplots(1, 2)
fig.set_figheight(7)
fig.set_figwidth(15)

x_ = np.arange(x_train.shape[1])

axes[0].hist(x=x_, bins=x_train.shape[1], weights=boosting.feature_importances_, color='blue')
lr_importances = abs(lr.coef_[0])
lr_importances = lr_importances / lr_importances.sum()
axes[1].hist(x=x_, bins=x_train.shape[1], weights=lr_importances, color='blue')

plt.show()
````

Получается, что дпри построении бустинга выделяются несколько важный признаков, а многи естальные имеют очень низкий показатель важности. Для логистической регрессии всё не так. Есть, конечно, немного выделяющиеся признаки, по важности, но в целом, всё коэффициенты достаточно близки по важности.

Кстати, чаще всего излишние признаки могут вредить качеству бустинга. Попробуйте отфильтровать на основании диаграммы хвост наименее важных признаков и снова обучить модель (с теми же гиперпараметрами). Стало ли лучше?

````python
mask = (boosting.feature_importances_ > 0.002)  # число 0.002 я подбирала так, чтобы качество всё же улучшилось
mask.sum()
````

Возьму только 60 лучших (согласно важности при построении градиентного бустинга) признаков, и построю модель только на этих признаках.

````python
best_features = np.arange(x_train.shape[1])[mask]
x_train_b = x_train[:, best_features]
x_valid_b = x_valid[:, best_features]
x_test_b = x_test[:, best_features]
````

````python
np.random.seed(111)

params = {'max_depth' : 6, 'min_samples_leaf' : 5, 'min_samples_split' : 15}
boosting_b = Boosting(base_model_params=params,
                      n_estimators=55,
                      learning_rate=0.7,
                      subsample=0.6)
boosting_b.fit(x_train_b, y_train, x_valid_b, y_valid)

train_score = boosting_b.score(x_train_b, y_train)
valid_score = boosting_b.score(x_valid_b, y_valid)
test_score = boosting_b.score(x_test_b, y_test)
    
print(f'train=%.4f' % train_score, f', val=%.4f' % valid_score, f', test=%.4f' % test_score)
````

При обучении бустинга с теми же параметрами, но только при обучении на 60-ти самых важных признаках, качество на тестовой выборке улушилось (было 0.9401, стало 0.9409).
